```sql
SELECT
    SUM(cnt) AS quarantine_row_count
FROM (
    SELECT COUNT(*) AS cnt FROM delta.`/Volumes/retail_catalog/retail_quarantine/q_sales_volume`
    UNION ALL
    SELECT COUNT(*) AS cnt FROM delta.`/Volumes/retail_catalog/retail_quarantine/q_customer_volume`
    UNION ALL
    SELECT COUNT(*) AS cnt FROM delta.`/Volumes/retail_catalog/retail_quarantine/q_product_volume`
) t;

```
