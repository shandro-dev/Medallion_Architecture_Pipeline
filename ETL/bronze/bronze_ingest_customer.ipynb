```python
from delta.tables import DeltaTable
from pyspark.sql.functions import *
```

```python
#paths
source_path = "/Volumes/retail_catalog/retail_source/source_customer_volume"
bronze_path = "/Volumes/retail_catalog/retail_bronze/bronze_customer_volume"
checkpoint_path="/Volumes/retail_catalog/retail_bronze/bronze_customer_volume/_checkpoint/"
schema_path="/Volumes/retail_catalog/retail_bronze/bronze_customer_volume/_schema/"
```

```python
df_sales=spark.readStream\
    .format("cloudFiles")\
    .option("cloudFiles.format","csv")\
    .option("header","true")\
    .option("cloudFiles.schemaLocation",schema_path)\
    .option("checkpointLocation",checkpoint_path)\
    .option("cloudFiles.includeExistingFiles", "false")\
    .load(source_path)
```

```python
df_sales.writeStream \
    .format("delta") \
    .outputMode("append") \
    .trigger(once=True) \
    .option("path", bronze_path) \
    .option("checkpointLocation", checkpoint_path) \
    .start() \
    .awaitTermination()
```
