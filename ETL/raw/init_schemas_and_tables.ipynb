```sql
%sql
-- 1️⃣ Create Catalog
CREATE CATALOG IF NOT EXISTS retail_catalog;

-- 2️⃣ Create Schemas for each layer
CREATE SCHEMA IF NOT EXISTS retail_catalog.retail_source;
CREATE SCHEMA IF NOT EXISTS retail_catalog.retail_bronze;
CREATE SCHEMA IF NOT EXISTS retail_catalog.retail_silver;
CREATE SCHEMA IF NOT EXISTS retail_catalog.retail_gold;
CREATE SCHEMA IF NOT EXISTS retail_catalog.retail_quarantine;

CREATE SCHEMA IF NOT EXISTS retail_catalog.audit;

-- 3️⃣ Create "Volumes" for each table (each volume corresponds to one table path)
-- Source Layer Volumes
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_source.source_product_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_source.source_customer_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_source.source_sales_volume;

-- Bronze Layer Volumes
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_bronze.bronze_product_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_bronze.bronze_customer_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_bronze.bronze_sales_volume;

-- Silver Layer Volumes
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_silver.silver_product_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_silver.silver_customer_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_silver.silver_sales_volume;

-- Gold Layer Volumes
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_gold.gold_sales_fact_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_gold.gold_product_dim_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_gold.gold_customer_dim_volume;

-- Quarantine Layer Volumes
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_quarantine.q_sales_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_quarantine.q_customer_volume;
CREATE VOLUME IF NOT EXISTS retail_catalog.retail_quarantine.q_product_volume;

-- audit
CREATE SCHEMA IF NOT EXISTS retail_catalog.audit;
CREATE VOLUME IF NOT EXISTS retail_catalog.audit.audit_volume;
-- Create the table inside the volume
CREATE TABLE IF NOT EXISTS delta.`/Volumes/retail_catalog/audit/audit_volume/etl_audit` (
    layer STRING,                  -- bronze / silver / gold
    table_name STRING,             -- e.g., bronze_customer
    load_time TIMESTAMP,           -- ETL run timestamp
    records_loaded BIGINT,         -- number of rows loaded in this run
    max_data_timestamp TIMESTAMP   -- max data_arrival_timestamp of rows loaded
)
USING DELTA;



DROP SCHEMA IF EXISTS retail_catalog.default CASCADE;
```
